```python
# Inspecting data for any NaN values.
# Number of rows with at least one NaN
num_rows_with_nan = agroFood_df.isna().any(axis=1).sum()
print(f"Number of rows with at least one NaN value: {num_rows_with_nan}")

# Number of rows where all values are NaN
num_rows_all_nan = agroFood_df.isna().all(axis=1).sum()
print(f"Number of rows where all values are NaN: {num_rows_all_nan}")

# Number of NaN values per row
nan_counts_per_row = agroFood_df.isna().sum(axis=1)
print("NaN counts per row:")
print(nan_counts_per_row)

# Number of NaN values per column
nan_counts_per_column = agroFood_df.isna().sum()
print("NaN counts per column:")
print(nan_counts_per_column)
```

```python
# Cleaning up Data removing columns that are not useful for our purposes.
AgroFood_drops_df = agroFood_df.drop(columns = ["Fires in humid tropical forests"], inplace = False);

#Running NaN check against AgroFood_drops for status update.
num_rows_with_nan = AgroFood_drops_df.isna().any(axis=1).sum()
print(f"Number of rows with at least one NaN value: {num_rows_with_nan}")

# Number of rows where all values are NaN
num_rows_all_nan = AgroFood_drops_df.isna().all(axis=1).sum()
print(f"Number of rows where all values are NaN: {num_rows_all_nan}")

# Number of NaN values per row
nan_counts_per_row = AgroFood_drops_df.isna().sum(axis=1)
print("NaN counts per row:")
print(nan_counts_per_row)

# Number of NaN values per column
nan_counts_per_column = AgroFood_drops_df.isna().sum()
print("NaN counts per column:")
print(nan_counts_per_column)
```
